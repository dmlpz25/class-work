## 1 - Toggle Line Numbers
Use the **View | Toggle Line Numbers** to turn line numbering on and off. While it has no impact on your Python code it can make referencing your code easier in your markdown explanations.

Example: Line 12 tests to make sure the number entered is within the acceptable range of 0 to 100.

You can try the other **View** menu options to see what they would do, the first two seem like personal preference items for non-mouse users. The last one offers some interesting possibilities for research documents **Cell Toolbar | None** will return you to the default.

```python
prompt = "Enter a number between 0 and 100 or 'done' when you are finished: "
nums = []
while True:
    test = input(prompt)
    if test.lower() == "done" : 
        break
    try:
        num = float(test)
    except:
        print("Invalid input. Please enter a number!")
        continue  
    if num < 0 or num > 100:
        print("Number out of range. Please enter a number between 0 and 100!")
        continue
    else:
        x = nums.append(num) 
if len(nums) > 0: 
    print("")
    for x in nums:
        if x >= 90:
            y = "A"
        elif x >= 80:
            y = "B" 
        elif x >= 70:
            y = "C"
        elif x >= 60:
            y = "D"
        else:
            y = "NC"
        print("The score", x, "equates to", y)  
else:
    print("")
    print("No numbers were entered! What's up with that Dude?")
```

## 3 - Hiding or Clearing Code Output
If you run the example code above with a few numbers you will see the results. You can toggle that cell's results off and on while in the cell with **Cell | Current Output | Toggle** (maybe to snag a screen capture before and after). To clear the output (maybe before saving your final results) use **Cell | Current Output | Clear**.

**Cell | All Outputs** offers the same choices for the entire notebook.

## 2 - Make a Copy
You can make a duplicate of any notebook with the **File | Make a Copy** feature. This is handy for carrying code from say a practice sheet to a document you want to save or use for another purpose. You can just delete any unwanted cells with the scissors button or **Edit | Cut Cell**. Note the new notebook has **-Copy1** appended to the name but **File | Rename** would let you change it.

**Note:** This creates a new notebook file immediately. If you decide to abandon it, you will need to delete it from your directory.

**Just a Reminder** - You can copy code from one cell to another using **[Ctrl]+c** and **[Ctrl]+v** or **Edit | Copy Cell**. This is hand when you want to make a progressing record of your coding process.

## 4 - Renaming a Notebook
In addition to using **File | Rename** tochange it, you can just clickon the file name at the top of the jupyter window.

## 5 - Things to look at when you have time
The **Edit** menu has some cool options for splitting and merging cell. **Edit | Split Cell** moves everything after the cursor to a new cell below.
**Help** has some useful info for Keyboard users as well as links the Help features for Notebook, Markdown, Python, and others.
**Help | User Interface Tour** is a simple tour of the notebook program.

## 6 - Using Notebook as a text Editor
From the original notebook tab you can see your Python.PY files listed with the notebooks. If you click on one, notebook will work as a simple editor with language specific color coding. It is not fancy, but if you just want to edit a script it works nicely. Just save your work when you are done and the script is up dated.

If you open the **Language** menu, you will see all the languages supported (with color coding).

## 7 - Adding Images to your Document
